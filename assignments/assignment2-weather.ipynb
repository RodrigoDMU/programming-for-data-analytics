```python
# Data frames. 
import pandas as pd
# Plotting.
import matplotlib.pyplot as plt
```

```python
# Load the Weather data set.
df = pd.read_csv("weatherreadings1.csv")
print (df)
```

The temperature was checked every 15 minutes for 1 day, starting on 02/10/2024 at 4:15pm and finishin on 03/10/24 at 3:00pm.
The column used  for x-axis was 'reportStartDateTime' and y-axis was 'dryBulbTemperature_Celsius'.

```python
# Informations of the dataset.
df.info()
```

```python
x = pd.to_datetime(df['reportStartDateTime']) # Convert argument to datetime.
y = df ['dryBulbTemperature_Celsius']

# Add figure size.
plt.figure(figsize = (14,8))

# Create a new figure and set of axes
plt.scatter(x, y, s = 8, color='red')
plt.plot(x, y, color='darkblue')

# Add grid
plt.grid(True)

# Add plot labels and title.
plt.title('Temperature Over Time (C)')
plt.xlabel('Report Start Date Time')
plt.ylabel('Dry Bulb Temperature (C)')

plt.show()
```
